#### Windows Users

Follow [How to Create SSH Keys with PuTTY on Windows](https://www.digitalocean.com/docs/droplets/how-to/add-ssh-keys/create-with-putty/)

#### For Windows 10 or 11
- Press Windows+R
- Enter cmd
- In the opened Command Prompt, type in “ssh-keygen”
- Press Enter
- You can choose to enter a passphrase, it will not be displayed.
- Go to the shown path to find your file named id_rsa.pub Ex. C:\Users\user/ .ssh/id_rsa.pub
- Open the file with a notepad and copy everything
- Go to github and click settings at top right 
- Go to SSH and GPG keys, click New SSH key and paste your SSH key here. 
- Click Add SSH key. You might be asked to enter Github password.
- Go back to your Command Prompt and type in “ssh -T git@github.com”
- Enter your SSH passphrase. 

- You will see “You’ve successfully authenticated” in the following message. Which means you are now connected to Github.

f) how would you make your local and remote copies change so that they have the most up-to-date version of the repo they are copied from?

To ensure that both your local and remote copies have the most up-to-date version of the repository they were copied from, you would:

Fetch the latest changes from the original source repository.

Update your remote repository with those changes.

Update your local repository to match the updated remote.

In practice, this can be done by:

Using git fetch to retrieve the latest changes from the original repository (often called upstream).

Using git merge or git rebase to integrate those changes.

Using git push to update your remote copy.

Using git pull (or git fetch + git merge) to update your local copy so it matches the remote.



g) why are there sometimes conflicts between copied repos / branches? How do you resolve them?

Conflicts happen when two copied repositories or branches modify the same file or same lines of code differently, and Git cannot automatically decide which version to keep.

To resolve conflicts, you manually edit the conflicted files, remove the conflict markers, choose or combine the correct changes, then run git add and complete the merge or rebase with git commit or git rebase --continue.

i) Write here some other commands we used in class and what they mean / how to use them:

git status — Shows what’s changed, what’s staged, and what branch you’re on.
Use: git status

git log — Shows commit history.
Use: git log (or shorter: git log --oneline)

git diff — Shows changes between versions.
Use:

Unstaged changes: git diff

Staged changes: git diff --staged

git add <file> — Stages a file for commit.
Use: git add notes.md (or stage everything: git add .)

git commit -m "message" — Saves staged changes as a commit.
Use: git commit -m "Update notes"

git pull — Fetches + merges changes from remote into your branch.
Use: git pull origin main (often better: git pull --rebase origin main)

git fetch — Downloads updates from remote without merging.
Use: git fetch origin

git push — Uploads your commits to the remote repo.
Use: git push origin main

git branch — Lists branches (or creates one).
Use:

List: git branch

Create: git branch new-branch

git checkout <branch> / git switch <branch> — Switches branches.
Use: git switch main

git checkout -b <branch> / git switch -c <branch> — Creates + switches to a new branch.
Use: git switch -c add-notes

git merge <branch> — Merges another branch into your current branch.
Use: git merge feature-branch

git rebase <branch> — Replays your commits on top of another branch (keeps history cleaner).
Use: git rebase main

git remote -v — Shows remote repo URLs (origin/upstream).
Use: git remote -v

git reset — Moves HEAD / unstages changes (can be destructive with --hard).
Use:

Unstage: git reset <file>

Undo commit but keep changes: git reset --soft HEAD~1

Match remote exactly: git reset --hard origin/main

git restore --staged <file> — Unstages a file (keeps file changes).
Use: git restore --staged notes.md
